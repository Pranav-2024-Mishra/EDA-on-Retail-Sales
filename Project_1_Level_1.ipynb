```python
Avg_price  = data.groupby('Product Category')['Price per Unit'].mean()
plt.figure(figsize =(8,5))
Avg_price.plot(kind='bar',color='green')
plt.title('Average Price per Unit by Product Category')
plt.xlabel('Product category')
plt.ylabel('Average Price')
plt.xticks(rotation=50)
```

```python
#Set the plot size
fig,ax = plt.subplots(figsize = (8,6))
sns.heatmap(Cov,annot=True,linewidths=0.99,
            cmap = 'bone',fmt =".2g")
plt.show()
```

```python
print(corr)
```

```python
#Set plot size
fig,ax = plt.subplots(figsize=(9,5))
sns.heatmap(corr,annot=True,linewidths=0.95,
            cmap='copper',fmt=".2g")
plt.show()
```

```python
data.head()
```

```python
#Set the figure size
plt.rcParams['figure.figsize']=[5,4]
sns.countplot(data=data,x='Product Category',color='blue')

#Title of the plot
plt.title('Count Plot of Product Category')
plt.xticks(rotation=50)
plt.show()
```

```python
#Barplot of the data
Total_amount = data['Total Amount'].value_counts()
sns.barplot(data=data,x='Product Category',y='Total Amount',color='r')
plt.title('Barplot for Produt Category and Total Amount')
plt.show()
```

```python
#Scatter plot
plt.scatter(data=data,x='Total Amount',y='Age',color='y')
plt.scatter(data=data,x='Quantity',y='Age',color='g')
plt.legend()
plt.show()
```

```python
#Histogram
data.plot.hist(subplots=True,figsize=(8,6),sharex=True,
               sharey=False,layout=(3,2))
plt.tight_layout()
plt.show()
```

```python
#Calculating length of the price per unit
Total_amount=data['Total Amount']
len(Total_amount)
```

```python
plt.rcParams['figure.figsize']=[8,3]
sns.distplot(Total_amount)
plt.title('Density Plot for Total Amount')
plt.show()
```

```python
quantity=data['Quantity']
len(quantity)
```

```python
plt.rcParams['figure.figsize']=[10,3]
sns.distplot(quantity,hist=False)
plt.title('Density plot for Quantity')
plt.show()
```

```python
#Violin plot
sns.violinplot(data=data,x='Total Amount',color='grey')
plt.title('Violin plot for Total Amount')
plt.show()
```

```python
plt.rcParams['figure.figsize']=[6,3]
sns.countplot(data = data, x='Gender',color ='yellow')
plt.title('Count plot for Age')
plt.show()
```

```python
#Pair plot
sns.pairplot(data=data,height=2.5)
plt.show()
```

```python
data['Product Category'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Product Category distribution')
plt.ylabel('')
plt.show()
```

```python
#Boxplot
data_numerical = data.describe(include=[np.number])
data_numerical.plot(kind='box',subplots=True,layout=(2,4))
plt.tight_layout()
plt.show()
```

```python
#Time Series analysis
sns.set(style='whitegrid')
plt.figure(figsize=(10,6))
sns.lineplot(data=data.head(), x='Date',y='Total Amount',color='grey')
plt.xlabel('Date')
plt.ylabel('Total Amount')
plt.title('Highest Amount Over the Time')
plt.show()
```
